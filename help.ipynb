```python
import eyeblue as eb
```

```python
### 行情API


# 获取ticker数据

# 参数： exchange(交易所):    详见json
#       symobol(货币类型):   详见json   

#返回：  字典类型 (嵌套)
#        cny: 对应人民币价格
#        open,close: 可能为空

eb.get_ticker(symbol='btc_usd', exchange='bitfinex')
```

```python
# 获取depth数据

# 参数： exchange(交易所):    详见json
#       symobol(货币类型):   详见json   

#返回：  字典类型 (嵌套)
#        aks: 卖
#        bis: 买
#        timestamp: 时间戳

eb.get_depth(symbol='btc_usd', exchange='bitfinex')
```

```python
# 获取K线数据

# 参数： exchange(交易所):    详见json
#       symobol(货币类型):   详见json   
#       k_type:   1,3,5,10,30min
#                 1day
#       size:     最大长度

#返回：  字典类型 (嵌套)
#        aks: 卖
#        bis: 买
#        timestamp: 时间戳

eb.get_k_data(symbol='btc_usd', k_type='1min', exchange='bitfinex',
          size=100)
```

```python
### 交易API

# 用户登陆

name, pwd = 1,1
u = eb.Util()
u.login(name,pwd)
```

```python
## 下单

# 参数： 
#       symobol(货币类型):   详见json   
#       price:       价格
#       amount:      数量
#       buy_sell：   委托方向, 0-买入，1-卖出
#       order_type： 委托类型（0-限价委托，1-市价委托）
#       open_close： 开平标识（0-开仓，1-平仓）
#       leverage：   期货合约杠杆，数字型字符串，比如 10 代表 10 倍杠杆。此字段对于
#                    现货交易不适用，直接留空



#返回：  
#       result:          0=失败，1=成功
#       error_code:     出错代码，仅当 result=0 时，此段才有值
#       error_message:  出错信息，仅当 result=0 时，此段才有值，最大长度为 50 字
#                       节
#       order_id:       委托序号，最大长度为 20 的字符串


#限价买入
re = u.order_send(symbol='btc_usdt',price=6002,amount=0.01,buy_sell=0,
                 order_type=0,open_close=0)
print(re)
```

```python
#市价卖出
re = u.order_send(symbol='btc_usdt',price=6002,amount=0.01,buy_sell=1,
                 order_type=1,open_close=0)
print(re)
```

```python
## 订单查询

#参数
#       symobol(货币类型):   详见json   
#       order_id：        委托序号，最大长度为 20 的字符串

#返回

#       result:            0=失败，1=成功
#       error_code:       出错代码，仅当 result=0 时，此段才有值
#       error_message:    出错信息，仅当 result=0 时，此段才有值，最大长度为 50 字节

#单条查询
re = u.order_query(symbol='btc_usdt',order_id='db4ee279-fe59-4e50-b859-2d34dec5a9ca')
print(re)
```

```python
#全部查询

re = u.order_query(symbol='btc_usdt',order_id=-1)
print(re)
```

```python
# 撤单
# buy_sell 委托方向, 0-买入，1-卖出，目前只有 bithumb 交易所要求提供此字段

re = u.order_kill(symbol='btc_usdt',order_id='db4ee279-fe59-4e50-b859-2d34dec5a9ca')
print(re)
```
